## Numbers on a Ring


# Regeln:

Die Zahlen 1..9 sollen so auf einen Ring gebracht werden, dass benachbarte Dreier-Tupel in Summe immer < 16 bleiben. Dieser Code soll herausfinden, ob es so eine Anordnung gibt und wenn ja wie sie lautet. 
Zunächst stellen wir fest: 7,8,9 müssen je zwei andere Zahlen zwischen sich haben, sonst wird die Regel direkt verletzt. Damit gibt es nur eine Möglichkeit, die drei Zahlen auf dem Ring anzubringen, alle anderen Möglichkeiten sind durch Drehung oder Spiegelung des Rings kongruent.
Wir definieren also eine Liste, die den Ring repräsentiert, d.h. im Kopf ist Ring[-1] nachbar von Ring[0]:

```python
Threshold = 16

ring = [9, None, None, 8, None, None, 7, None, None]
print(ring)
```

Es verbleiben die Zahlen 1..6, die wir jetzt durch-permutieren wollen. Dazu eine zweite Liste

```python
small_numbers = list(range(1,7))
print(small_numbers)
```

```python
for number1 in small_numbers:
    ring[1] = number1
    remaining1 = small_numbers.copy()
    remaining1.remove(number1)
    # print(remaining1)
    for number2 in remaining1:
        ring[2] = number2
        remaining2 = remaining1.copy()
        remaining2.remove(number2)
        # print(remaining2)
        # print(remaining1)
        for number4 in remaining2:
            ring[4] = number4
            remaining4 = remaining2.copy()
            remaining4.remove(number4)
            for number5 in remaining4:
                ring[5] = number5
                remaining5 = remaining4.copy()
                remaining5.remove(number5)
                for number7 in remaining5:
                    ring[7] = number7
                    remaining7 = remaining5.copy()
                    remaining7.remove(number7)
                    for number8 in remaining7:
                        ring[8] = number8
                        iteration = True
                        for index in range(-1,8):
                            # print(index)
                            if ring[index-1] + ring[index] + ring[index+1] >= Threshold:
                                iteration = False
                        if iteration == True:
                            print('\x1b[6;30;42m' + "Found a working Permutation!" + '\x1b[0m')
                            print(ring)
                        else:
                            print("--> Rule violated!")
                            print(ring)
```
